```python
import numpy as np
```

```python
points = np.array([
    [0.1,0.6],[0.15,0.71],[0.08,0.9],[0.16,0.85],
    [0.2,0.3],[0.25,0.5],[0.24,0.1],[0.3,0.2]
    ])
```

```python
m1=points[0]
m2=points[7]
```

```python
def dist(p1,p2) : return np.sqrt(np.sum((p1-p2)**2))
```

```python
cluster1,cluster2=[],[]
```

```python
for p in points:
  if dist(p,m1)<dist(p,m2):
    cluster1.append(p)
  else:
    cluster2.append(p)
```

```python
cluster1=np.array(cluster1)
cluster2=np.array(cluster2)
```

```python
new_m1=np.mean(cluster1,axis=0)
new_m2=np.mean(cluster2,axis=0)
```

```python
print("Cluster 1 Points:\n", cluster1)
print("Cluster 2 Points:\n", cluster2)
print("Updated m1:", new_m1)
print("Updated m2:", new_m2)
```
