```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
df=pd.read_csv('bank_transactions_data_2.csv')
df.head()
```

```python
df=df.drop(columns=['TransactionID','AccountID','TransactionDate','DeviceID','IP Address','MerchantID','PreviousTransactionDate'],axis=1)
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.dtypes
```

```python
df.head()
```

```python
le=LabelEncoder()
for col in df.select_dtypes(include="object"):
  df[col]=le.fit_transform(df[col])
```

```python
X=df.drop('TransactionType',axis=1)
y=df['TransactionType']
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.25,random_state=42)
```

```python
#Linear
model_linear=SVC(kernel='linear')
model_linear.fit(X_train,y_train)
y_pred_linear=model_linear.predict(X_test)
print(accuracy_score(y_test,y_pred_linear))
print(confusion_matrix(y_test,y_pred_linear))
```

```python
#Polynomial
model_poly=SVC(kernel='poly',degree=3)
model_poly.fit(X_train,y_train)
y_pred_poly=model_poly.predict(X_test)
print(accuracy_score(y_test,y_pred_poly))
print(confusion_matrix(y_test,y_pred_poly))
```
